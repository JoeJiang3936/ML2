```python
import pandas as pd
import numpy as np
df = pd.read_csv('claim.sample.csv', delimiter = ',', na_values=[' '])
#if you go back to the csv file, all missing cells have one space in them, initially I was really buffled by no missing values 
```

```python
df.shape
```

```python
#data types for each columns and missing data 
df.info()
```

```python
#subset the dataframe to get the J-code claims
df = df[df['Procedure.Code'].notna()]
J_claims_df = df[df['Procedure.Code'].str.startswith('J')]
```

```python
#number of J-code claims
J_claims_df.shape
```

```python
#generate the labels 
J_claims_df['Provider.Payment.Amount'] = J_claims_df['Provider.Payment.Amount'] > 0 
J_claims_df['Provider.Payment.Amount'] = J_claims_df['Provider.Payment.Amount'].astype(int)
```

```python
#checking for missing values
J_claims_df.isnull().sum()
```

```python
#checking for levels in each columns
for c in J_claims_df.columns:
    print(c, len(J_claims_df[c].unique()))
```

```python
#from above, dropping some columns that are likely not useful for the model... my own educated guess here
J_claims_df = J_claims_df.drop(columns = ['V1', 'Claim.Number', 'Claim.Line.Number', 'Member.ID', 'Place.Of.Service.Code', 'Group.Index', 'Subscriber.Index'])
```

```python
#some overall stats of J_claims_df, you can spend a lot of time on this if you want
import pandas_profiling
pandas_profiling.ProfileReport(J_claims_df)
```

```python
#from the profiling, I will drop Subscriber.Payment.Amount column because there is only one claim has a value different from 0.
J_claims_df = J_claims_df.drop(columns = ['Subscriber.Payment.Amount'])
```

```python
#check the datatype again
J_claims_df.info()
```

```python
#Cast the following columns as categorical
for c in ['Line.Of.Business.ID','Subgroup.Index', 'Claim.Current.Status']:
    J_claims_df[c] = J_claims_df[c].astype(str) 
```

```python
#Impute two columns, Revenue.Code and Service.Code, with their mode 
for column in ['Revenue.Code', 'Service.Code']:
    J_claims_df[column].fillna(J_claims_df[column].mode()[0], inplace=True)
```

```python
#fill the rest NAs with 'None' as its own category 
J_claims_df.fillna('None', inplace = True)
```

```python
#check the df for missing value again
J_claims_df.isnull().sum()
```

```python
#Train and test data split, data exploration is done using train data set to avoid data-snooping 
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(J_claims_df, test_size=0.1)
```

```python
##Cross-tabulation between columns with lots of missing value and the label
for column in ['Denial.Reason.Code', 'Line.Of.Business.ID', 'Price.Index', 'In.Out.Of.Network', 'Reference.Index',
               'Pricing.Index', 'Capitation.Index', 'Claim.Pre.Prince.Index', 'Network.ID', 'Agreement.ID']:
    print(pd.crosstab(train_df[column], train_df['Provider.Payment.Amount'], margins=True))
    print('\n')
```

```python
#Paid claims by subgroup index (~ 800 levels)
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize= (40,40))
data_grouped = train_df.groupby(by=train_df['Subgroup.Index'])
paid_rate = data_grouped['Provider.Payment.Amount'].sum()/data_grouped['Provider.Payment.Amount'].count()
paid_rate = paid_rate.sort_values()
paid_rate.plot(kind='barh')
plt.title('paid Rate by subgroups')
plt.show()
```

```python
#Paid claims by diagnosis code (>1400 levels)
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize= (40,40))
data_grouped = train_df.groupby(by=train_df['Diagnosis.Code'])
paid_rate = data_grouped['Provider.Payment.Amount'].sum()/data_grouped['Provider.Payment.Amount'].count()
paid_rate = paid_rate.sort_values()
paid_rate.plot(kind='barh')
plt.title('paid Rate by diagnosis code') 
plt.show()
```

```python
#checking for levels in each columns
for c in train_df.columns:
    print(c, len(train_df[c].unique()))
```

```python
train_df.shape
```

```python
#labels
labels = train_df['Provider.Payment.Amount']
```

```python
#one hot encoding for subgroup.index column
data_train = pd.get_dummies(train_df['Subgroup.Index'])
```

```python
data_train.shape
```

```python
#Selecting features using random Forest model
from sklearn.ensemble import RandomForestClassifier

#split into train and valid datset
X_train, X_valid, y_train, y_valid = train_test_split(data_train, labels, test_size=0.10, random_state=101)

#train a model
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train,y_train)

#Variable importance table
feature_imp = pd.Series(rf.feature_importances_,index=X_train.columns).sort_values(ascending=False)
```

```python
#top 40 features
feature_imp[:40].index
```

```python
#make prediction of test dataset
from sklearn.metrics import *
predictions_randf = rf.predict(X_valid)
auc = roc_auc_score(y_valid, predictions_randf)
auc
```
